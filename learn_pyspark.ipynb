```
from pyspark.sql import SparkSession
```

```
spark = SparkSession.builder.appName('demo').getOrCreate()
```

```
spark
```

```
url = "https://www.stats.govt.nz/assets/Uploads/Effects-of-COVID-19-on-trade/Effects-of-covid-19-on-trade-at-28-april-2021-provisional/Download-data/effects-of-covid-19-on-trade-at-28-April-2021-provisional.csv"
```

```
from pyspark import SparkFiles
```

```
spark.sparkContext.addFile(url)
```

```
df = spark.read.csv("file://"+SparkFiles.get("effects-of-covid-19-on-trade-at-28-April-2021-provisional.csv"), header=True, inferSchema= True)
```

```
df.show()
```

```
df
```

```
df.count()
```

```
df.createOrReplaceTempView("tableA")
spark.sql("SELECT count(*) from tableA").show()
```
